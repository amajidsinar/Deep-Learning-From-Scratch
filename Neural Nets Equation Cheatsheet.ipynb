# Every equations of neural nets

Error of neuron in layer L

\begin{align}
\delta^L = \frac{\partial C}{\partial z^L}
\end{align}

Error of neuron in the output layer relative to activated output (in equation form) -- for sigmoid layer
\begin{align}
\delta^L = \frac{\partial C}{\partial a^L} \sigma'(z^L)
\end{align}


Backpropagation (with SGD)
\begin{align}
w \Rightarrow w' = w - \frac{\eta}{m} \frac{\partial C}{\partial w} \\
b \Rightarrow b' = b - \frac{\eta}{m} \frac{\partial C}{\partial b}
\end{align}


Error of neuron in terms of the error in the previous layer -- for sigmoid layer
\begin{align}
\delta^{L-1} = ((w^L)^T)\delta^L \odot \sigma'(z^{L-1})
\end{align}

Error of neuron in terms of the error in the previous layer -- for softmax layer
\begin{align}
\delta^{L-1} = (y-a^L)(w^L)(a^{L-1}(1-a^{L-1}))
\end{align}

Rate of change of the cost with respect to any weight
\begin{align}
\frac{\partial C}{\partial w^L} = a^{L-1}\partial^L
\end{align}

Rate of change of the cost with respect to any bias

\begin{align}
\frac{\partial C}{\partial b^L} = \partial^L
\end{align}


# Cost function

Mean Square Error

\begin{align}
C(w,b)=\frac{1}{2n}\sum|y-a|^2 \\
\frac{\partial C}{\partial a^L} = (a^L - y)
\end{align}

Log loss
\begin{align}
C = -ln a^L \\
\frac{\partial C}{\partial a^L} = -\frac{1}{a^L}
\end{align}

Cross entropy
\begin{align}
C=-\frac{1}{n}\sum[ylna+1-yln(1-a] \\
If using sigmoid layer to xent
\frac{\partial C}{\partial a^L} = \frac{a-y}{a(a-a)}
\end{align}

# Output layer

Sigmoid
\begin{align}
a^L = \sigma(w^L a^{L-1})
\end{align}

However with sigmoid output layer you will have to calculate $\frac{\partial C}{\partial a^L}$

Softmax
\begin{align}
a^L = \frac{e^{{z}^{L}}} {\sum e^{{z}^{L}}}  \\
\delta^L = y - a^L
\end{align}

which means that, when using softmax layer, we are not concerned with $\sigma'(z^L)$

Up until this step, every kind of layers, activation functions and everything else is the same. However, every pair of cost function and its corresponding output layer have different $\delta^L$ error

Error of neuron in the output(last) layer relative to activated output (in matrix form)

\begin{align}
\delta^L
\end{align}

Error of neuron in the L layer relative to $L^{-1}$ (in matrix form)

\begin{align}
\delta^{L-1}
